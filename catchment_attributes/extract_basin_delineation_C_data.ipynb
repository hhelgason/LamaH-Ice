```python
from pathlib import Path
import pandas as pds
import geopandas as gpds
```

```python
# Basin delineation C only includes data from catchments with no or low influence
```

```python
# We process:
# Catchment attributes (from delineation B)
# gauge and basins shapefiles
# gauge hierarchy file
# Lastly: Timeseries (daily discharge and daily meteorological forcings)
```

```python
basins_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\B_basins_intermediate_all\3_shapefiles\Basins_B.shp")
gauges_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\D_gauges\3_shapefiles\gauges.shp")
attrs_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\B_basins_intermediate_all\1_attributes\Catchment_attributes.csv")
hier_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\B_basins_intermediate_all\1_attributes\Gauge_hierarchy.csv")

new_basins_path =  Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\C_basins_intermediate_lowimp\3_shapefiles\Basins_C.shp")
new_attrs_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\C_basins_intermediate_lowimp\1_attributes\Catchment_attributes.csv")
new_hier_path = Path(r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\C_basins_intermediate_lowimp\1_attributes\Gauge_hierarchy.csv")

# Read the gauges shapefile
gauges_gdf = gpds.read_file(gauges_path)
gauges_gdf = gauges_gdf.set_index('id')

# Read the watersheds shapefile 
basins_gdf = gpds.read_file(basins_path)
basins_gdf = basins_gdf.set_index('id')

# Read the attributes csv file
attrs = pds.read_csv(attrs_path)
attrs = attrs.set_index('id')

# Read the gauge hierarchy csv file
hier = pds.read_csv(hier_path,sep=';')
hier = hier.set_index('ID')

# We only want to keep data from gauges with no or low impact (degimpact as "l" or "u")
masker = gauges_gdf[gauges_gdf['degimpact'].isin(['l', 'u'])].index
# Since we do not currently have data for splitted gauges, we skip the only splitted gauge with no influence (990)
######################################## To do: Remove this once the data has been computed #####################################################################################################################
masker_attrs = masker.sort_values()[:-1]

# We extract only the desired data from basins_gdf, attrs and hier and save using the new paths defined above
new_basins_gdf = basins_gdf.loc[masker]
new_attrs = attrs.loc[masker_attrs]
new_hier = hier.loc[masker]

# Save the extracted data to the new paths
new_basins_gdf.to_file(new_basins_path)
new_attrs.to_csv(new_attrs_path)
new_hier.to_csv(new_hier_path, sep=';')
```

# Now we process meteorological timeseries

```python
import os
import shutil

source_directory = r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\B_basins_intermediate_all\2_timeseries\daily\meteorological_data"
destination_directory = r"C:\Users\hordurbhe\Documents\Vinna\lamah\lamah_ice\lamah_ice\C_basins_intermediate_lowimp\2_timeseries\daily\meteorological_data"

# List of file name numbers
file_numbers = list(masker)  # Add the numbers you want

for number in file_numbers:
    source_file = os.path.join(source_directory, f"ID_{number}.csv")
    destination_file = os.path.join(destination_directory, f"ID_{number}.csv")
    
    if os.path.exists(source_file):
        shutil.copy(source_file, destination_file)
        print(f"File {number}.csv copied successfully.")
    else:
        print(f"File {number}.csv not found in the source directory.")
```
